```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
# Read data from a csv
dataset = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')
dataset.head()
```

```python
fig = go.Figure(data=[go.Surface(z=dataset.values)])
fig.update_layout(title='Mt Bruno Elevation')
fig.update_layout(template='plotly_dark')
fig.show()
```
