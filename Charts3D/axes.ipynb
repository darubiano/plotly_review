```python
import plotly.graph_objects as go
import numpy as np
```

```python
N = 70
fig = go.Figure(data=[go.Mesh3d(x=(70*np.random.randn(N)),
                y=(55*np.random.randn(N)),
                z=(40*np.random.randn(N)),
                opacity=0.5,
                color='rgba(244,22,100,0.6)')])
fig.update_layout(template='plotly_dark')
fig.show()
```

```python
N = 50
fig1 = go.Figure(data=[go.Mesh3d(x=(30*np.random.randn(N)),
                y=(25*np.random.randn(N)),
                z=(30*np.random.randn(N)),
                opacity=0.5,color='white')])
fig1.update_layout(scene = dict(
                    xaxis = dict(
                    backgroundcolor="rgb(0,120,255)",
                    gridcolor="white",
                    showbackground=True,
                    zerolinecolor="white",),
                    yaxis = dict(
                        backgroundcolor="rgb(203,50,52)",
                        gridcolor="white",
                        showbackground=True,
                        zerolinecolor="white"),
                    zaxis = dict(
                        backgroundcolor="rgb(212,175,55)",
                        gridcolor="white",
                        showbackground=True,
                        zerolinecolor="white",),),
                    margin=dict(
                    r=10, l=10,
                    b=10, t=10))
fig1.update_layout(template='plotly_dark')                    
fig1.show()
```
